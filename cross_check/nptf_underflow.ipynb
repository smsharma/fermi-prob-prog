```python
npix = 1000

theta = jnp.array([[2, 5., 1.3, -5.4, 10., 4.]])
pt_sum_compressed = 1.3 * jnp.ones((npix,))
npt_compressed = jnp.ones((1, npix))
data = jnp.arange(npix)
f_ary = m.f_ary
df_rho_ary = m.df_rho_ary
k_max = npix - 1
npixROI = npix

ll_s = log_like_np(theta, pt_sum_compressed, npt_compressed, data, f_ary, df_rho_ary, k_max, npixROI)
nan_mask = jnp.isnan(ll_s)
nan_idx = np.where(np.asarray(nan_mask))[0]
print(ll_s)
if len(nan_idx) > 0:
    print(int(nan_idx[0]))
```
