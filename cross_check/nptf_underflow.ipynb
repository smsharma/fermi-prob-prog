```python
%reload_ext autoreload
%autoreload 2

import os
import sys
import numpy as np
import jax
jax.config.update("jax_enable_x64", True)
import jax.numpy as jnp
import healpy as hp
```

```python
from fpp.likelihoods.npll_jax_log import log_like_np
from fpp.likelihoods.npll_jax_old import log_like_np as log_like_np_old
```

```python
from fpp.models.np_model import NPModel
m = NPModel(data=np.ones((hp.nside2npix(128),)))
```

```python
n_pix = 500

theta = jnp.array([[10, 5., 1.3, -5.4, 10., 4.]])
pt_sum = 13 * jnp.ones((n_pix,))
npt = jnp.ones((1, n_pix))
data = jnp.arange(n_pix)
f_ary = m.f_ary
df_rho_ary = m.df_rho_ary
k_max = n_pix - 1

ll_s = log_like_np(theta, pt_sum, npt, data, f_ary, df_rho_ary, k_max, n_pix)
ll_s_old = log_like_np_old(theta, pt_sum, npt, data, f_ary, df_rho_ary, k_max, n_pix)

print("New LL:", ll_s)
print("Old LL:", ll_s_old)
print("Relative difference:", jnp.abs((ll_s - ll_s_old) / ll_s))
```
