# 2. Renormalize

```python
var_names = ['S_iso', 'S_bub', 'S_psc', 'S_pib', 'S_ics', 'S_nfw', 'Sps_nfw', 'n1_nfw', 'n2_nfw', 'sb1_nfw', 'Sps_dsk', 'n1_dsk', 'n2_dsk', 'sb1_dsk']
```

```python
ratio_d = {
    'S_pib': np.mean(m.pib[~m.mask_roi]) / np.mean(m.pib[~m.mask_plane]),
    'S_ics': np.mean(m.ics[~m.mask_roi]) / np.mean(m.ics[~m.mask_plane]),
    'S_bub': np.mean(m.bub[~m.mask_roi]) / np.mean(m.bub[~m.mask_plane]),
    'S_psc': np.mean(m.psc[~m.mask_roi]) / np.mean(m.psc[~m.mask_plane]),
    'S_iso': np.mean(m.iso[~m.mask_roi]) / np.mean(m.iso[~m.mask_plane]),
    'S_nfw': np.mean(m.nfw_1p0[~m.mask_roi]) / np.mean(m.nfw_1p0[~m.mask_plane])
}
```

```python
tag = "test_0123.09"
chain_file = f"/n/home07/yitians/fermi/NPTFit/examples/chains/{tag}/post_equal_weights.dat"
samples = np.array(np.loadtxt(chain_file)[:, :-1])
d = {}
for i, t in enumerate(var_names):
    if t in ratio_d:
        d[t] = samples[:, i] * ratio_d[t]
    else:
        d[t] = samples[:, i]
pickle.dump(d, open(f"/n/home07/yitians/fermi/fermi-prob-prog/outputs/fits_cmp/nptfit/{tag}_samples_shifted.p", "wb"))
```
